```python
import pandas as pd
import numpy as np

## For plotting
import matplotlib.pyplot as plt
import seaborn as sns

## This sets the plot style
## to have a grid on a white background
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Imputed_data_Ivan.csv")
df
```

```python
X = df.iloc[:,0:50]
y = df.iloc[:,51]  # Target is the Parkinsons column
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                test_size=.25,
                                                shuffle=True,
                                                random_state=614,
                                                stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
## Accuracy
from sklearn.metrics import accuracy_score
## Precision
from sklearn.metrics import precision_score
## Recall
from sklearn.metrics import recall_score
```

```python
n=10
for j in range(1,n):
    forest_clf = RandomForestClassifier(n_estimators = 900,max_depth=j, random_state = 614)
    forest_clf.fit(X_train,y_train)
    y_pred_train = forest_clf.predict(X_train)
    accs_train = accuracy_score(y_train,y_pred_train)
    precs_train = precision_score(y_train,y_pred_train)
    recs_train= recall_score(y_train,y_pred_train)
    
```

```python
print(accs_train)
print(precs_train)
print(recs_train)
```

```python
y_pred_test = forest_clf.predict(X_test)
accs_test = accuracy_score(y_test,y_pred_test)
precs_test = precision_score(y_test,y_pred_test)
recs_test= recall_score(y_test,y_pred_test)
```

```python
print(accs_test)
print(precs_test)
print(recs_test)
```

```python
forest_clf.feature_importances_
```

```python
names = []
scores = []
for name, score in zip(X_train.columns,forest_clf.feature_importances_):
    names.append(name)
    scores.append(np.round(score,4))
    
score_df = pd.DataFrame({'feature':names,'importance_score':scores})

score_df.sort_values('importance_score',ascending=False)
```

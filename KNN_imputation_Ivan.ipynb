```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("ConvertedData-All.csv")
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer=KNNImputer(n_neighbors=4)
```

```python
df_times=df.iloc[:,14:65]
df_times
```

```python
df_times.isnull().sum()
```

```python
df_noNaN=imputer.fit_transform(df_times)
df_noNaN
```

```python
pd_imputed=pd.DataFrame(df_noNaN)
pd_imputed
```

```python
pd_imputed.to_csv(r'Imputed_data_Ivan.csv', index = False)
```
